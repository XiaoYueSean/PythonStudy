```python
import os
import json
json_folder_path = '/Users/sean/Desktop/form_understanding/1data/airwayBill_test/ccint'
ccint_folder = os.listdir(json_folder_path)

for files in ccint_folder:
    json_path = os.path.join(json_folder_path, files)
    try:
        aws = json.dumps({'Blocks': get_blocks(json_path)})
        with open(os.path.join('/Users/sean/Desktop/form_understanding/1data/airwayBill_test/newccint', files), 'w') as f:
            f.write(aws)
    except:
        pass
    
```

```python
def get_blocks(json_file_path):
    file = open(json_file_path,'rt',encoding='utf-8')
    json_str = file.read()
    ocr_obj = json.loads(json_str)

    def average(seq):
        return float(sum(seq)) / len(seq)

    rslt = ocr_obj["result"]
    image_width = rslt["rotated_image_width"]
    image_height = rslt["rotated_image_height"]

    aws_blocks = []
    for line in rslt["lines"]:
        p = line['position']
        minx = min(p[0],p[6])
        miny = min(p[1],p[3])
        maxx = max(p[2],p[4])
        maxy = max(p[5],p[7])

        Block = {}
        Block['BlockType'] = "LINE"
        Block['Text'] = line['text']
        Block['Confidence'] = line['score']
        Block['Geometry'] = {
                    "BoundingBox": {
                        "Width": (maxx-minx)/image_width,
                        "Height": (maxy-miny)/image_height,
                        "Left": minx/image_width,
                        "Top": miny/image_height
                    },
                    "Polygon": [
                        {
                            "X": p[0]/image_width,
                            "Y": p[1]/image_height
                        },
                        {
                            "X": p[2]/image_width,
                            "Y": p[3]/image_height
                        },
                        {
                            "X": p[4]/image_width,
                            "Y": p[5]/image_height
                        },
                        {
                            "X": p[6]/image_width,
                            "Y": p[7]/image_height
                        }
                    ]
        }

        aws_blocks.append(Block)

    #trans for WORD
    for line in rslt["lines"]:
        line_txt = line["text"].replace("　"," ").strip()
        if " " in line_txt:
            #拆分WORD
            start_pos = 0
            for word in line_txt.split(" "):
                Block = {}
                Block['BlockType'] = "WORD"
                Block['Text'] = word
                start_char_index = line_txt.find(word,start_pos)
                end_char_index_p = start_char_index + len(word)
                end_char_index = end_char_index_p - 1
                start_pos = end_char_index_p
                Block['Confidence'] = average(line['char_score'][start_char_index:end_char_index_p])

                cp = line['char_polygons']
                p[0] = cp[start_char_index][0]
                p[1] = cp[start_char_index][1]
                p[2] = cp[end_char_index][2]
                p[3] = cp[end_char_index][3]
                p[4] = cp[end_char_index][4]
                p[5] = cp[end_char_index][5]
                p[6] = cp[start_char_index][6]
                p[7] = cp[start_char_index][7]

                minx = min(p[0],p[6])
                miny = min(p[1],p[3])
                maxx = max(p[2],p[4])
                maxy = max(p[5],p[7])

                Block['Geometry'] = {
                            "BoundingBox": {
                                "Width": (maxx-minx)/image_width,
                                "Height": (maxy-miny)/image_height,
                                "Left": minx/image_width,
                                "Top": miny/image_height
                            },
                            "Polygon": [
                                {
                                    "X": p[0]/image_width,
                                    "Y": p[1]/image_height
                                },
                                {
                                    "X": p[2]/image_width,
                                    "Y": p[3]/image_height
                                },
                                {
                                    "X": p[4]/image_width,
                                    "Y": p[5]/image_height
                                },
                                {
                                    "X": p[6]/image_width,
                                    "Y": p[7]/image_height
                                }
                            ]
                }
            aws_blocks.append(Block)
        else:
            p = line['position']
            minx = min(p[0],p[6])
            miny = min(p[1],p[3])
            maxx = max(p[2],p[4])
            maxy = max(p[5],p[7])

            Block = {}
            Block['BlockType'] = "WORD"
            Block['Text'] = line['text']
            Block['Confidence'] = line['score']
            Block['Geometry'] = {
                        "BoundingBox": {
                            "Width": (maxx-minx)/image_width,
                            "Height": (maxy-miny)/image_height,
                            "Left": minx/image_width,
                            "Top": miny/image_height
                        },
                        "Polygon": [
                            {
                                "X": p[0]/image_width,
                                "Y": p[1]/image_height
                            },
                            {
                                "X": p[2]/image_width,
                                "Y": p[3]/image_height
                            },
                            {
                                "X": p[4]/image_width,
                                "Y": p[5]/image_height
                            },
                            {
                                "X": p[6]/image_width,
                                "Y": p[7]/image_height
                            }
                        ]
            }
            aws_blocks.append(Block)
    return  aws_blocks
```
